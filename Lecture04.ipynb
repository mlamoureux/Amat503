```python
A = zeros((100,100))
```

```python
imshow(A, cmap='gray')
```

```python
for i in range(50):
    for j in range(50):
        A[i,j] = 0.5
        
```

```python
imshow(A, cmap='gray')
```

```python
for i in range(100):
    for j in range(100):
        if ((i-50)**2 + (j-50)**2 < 100):
            A[i,j] = 1.0
```

```python
imshow(A, cmap='gray')
```
